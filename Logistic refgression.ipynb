```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import average_precision_score
```

```python
data = pd.read_excel('Book1.xlsx')
data
```

```python
# data.dropna(inplace=True,axis=1, how='all')
data.dropna(inplace=True)
```

```python
#fiktering the people with corona
data = data[data['COVID-19 Status \n(Positive)']==1]
data
```

```python
data.columns
```

```python
col = [ 'Gender','Non Invasive Positive Pressure Ventilation (NIPPV) ',
       'Mechanical Ventilation', 'ECMO', 'HFNC', 'Received tracheostomy',
       'Prone-positioning ventilation', 'Vasopressors', 'TPE', 'Remdesivir',
       'Hydroxychloroquine', 'Antibacterial agent', 'Corticosteroid',
       'Interleukin-6 receptor antagonist',
       'Renal Replacement Therapy (Dialysis) in ICU',
       'Chronic Cardiovascular-Lung Disease', 'Chronic Liver Kidney Disease',
       'Diabetes', 'Hypertension', 'Any other Risk Factors or Comorbidities?']
```

```python
x = pd.DataFrame(data , columns=col)
x = x.astype(int)
x
```

```python
y = pd.DataFrame(data , columns=['ICU Outcome'])
y = y.astype(int)
y['ICU Outcome'].value_counts()
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size = 0.2,random_state=10)
print('x [features] train data', x_train.shape)
print('y [respond ] train data',y_train.shape)
print('x [features] Test data',x_test.shape)
print('y [respond ] Test data',y_test.shape)
```

```python
clf = LogisticRegression().fit(x_train, y_train)
p_train = clf.predict(x_train)
acc1 = accuracy_score(y_train, p_train)
print('Accuracy score for train : {0:0.2f}'.format(acc1))
```

```python
p_test = clf.predict(x_test)
acc2 = accuracy_score(y_test, p_test)
print('Accuracy score for test : {0:0.2f}'.format(acc2))
```

```python
average_precision = average_precision_score(y_train, p_train)
print('Average precision-recall score for train : {0:0.2f}'.format(average_precision))
```

```python
average_precision = average_precision_score(y_test, p_test)
print('Average precision-recall score fro test : {0:0.2f}'.format(average_precision))
```

```python
y_test['predicted value']= p_test

y_test.replace(1.0,'Yes',inplace=True)
y_test.replace(0.0,'No',inplace=True)

y_test
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf.fit(x, y), x, y,
                              cv=5,
                              scoring='accuracy')

print("Accuracy scores:\n", scores)
print(scores.mean())
```
